```python
import torch
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
import matplotlib.pyplot as plt
```

```python
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])
```

```python
dataset = ImageFolder('Datasets/dataset_20_classes/', transform=transform)
```

```python
class_mean = []
class_std = []
for i in range(len(dataset.classes)):
    class_mean.append(0)
    class_std.append(0)
```

```python
for image, label in dataset:
    class_mean[label] += image.mean()
    class_std[label] += image.std()
```

```python
for i in range(len(dataset.classes)):
    class_mean[i] /= dataset.targets.count(i)
    class_std[i] /= dataset.targets.count(i)
```

```python
import numpy as np
```

```python
fig, ax = plt.subplots(2, figsize=(10, 8))
ax[0].bar(range(len(dataset.classes)), class_mean, yerr=0, align='center', alpha=0.5, ecolor='black', capsize=10)
ax[0].set_xticks(range(len(dataset.classes)))
ax[0].set_xticklabels(dataset.classes, rotation=90)
ax[0].set_ylabel('Pixel Value')
ax[0].set_title('Mean Pixel Values for 20 class dataset ')

ax[1].bar(range(len(dataset.classes)), class_std, yerr=0, align='center', alpha=0.5, ecolor='black', capsize=10)
ax[1].set_xticks(range(len(dataset.classes)))
ax[1].set_xticklabels(dataset.classes, rotation=90)
ax[1].set_ylabel('Pixel Value')
ax[1].set_title('Standard Deviation of Pixel Values for 20 class dataset')

plt.tight_layout()
plt.show()
```
